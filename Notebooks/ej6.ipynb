Procesamiento de hashtags

En esta celda:

```python
todos_los_hashtags = obtener_todos_los_hashtags(posts)
frecuencias = contar_frecuencias(posts)
trending = obtener_trending(frecuencias)
total_unicos = total_hashtags_unicos(frecuencias)
```

```python
print("Hashtags trending (más de una aparición):\n")
for hashtag, cantidad in trending.items():
    print(f"{hashtag}: {cantidad}")
print("\nTotal de hashtags únicos:", total_unicos)
```
