Importación de funciones

En esta celda importamos desde la carpeta `src` las funciones necesarias para calcular el costo de envío.
Estas funciones permiten:
-verificar si la zona ingresada es válida,
-determinar el rango de peso del paquete,
-y calcular el costo de envío según el peso y la zona.

```python
import sys
import os
sys.path.append(os.path.abspath(".."))
from src.ejercicio5 import calcular_costo_envio, zona_valida
```

Ingreso de datos

En esta celda solicitamos al usuario:
-el peso del paquete en kilogramos,
-y la zona de destino (local, regional o nacional).

```python
peso = float(input("Ingrese el peso del paquete (kg): "))
zona = input("Ingrese la zona de destino (local/regional/nacional): ").lower()
```

Validación y cálculo

En esta celda:
-verificamos si la zona es válida,
-y si lo es, calculamos el costo de envío según el peso y la zona.

```python
if not zona_valida(zona):
    print("Zona no válida. Las zonas disponibles son: local, regional, nacional.")
else:
    costo = calcular_costo_envio(peso, zona)
    print("Costo de envío: $" + str(costo))
```
