```python
review = """La película sigue a un grupo de astronautas que
viajan a Marte
en una misión de rescate. El capitán Torres lidera al equipo
a través
de tormentas solares y fallos en el sistema de navegación. Al
llegar
a Marte descubren que la base está abandonada y los
suministros
destruidos. Torres decide sacrificar la nave nodriza para
salvar
al equipo y logran volver a la Tierra en una cápsula de
emergencia.
El final revela que Torres sobrevivió gracias a un pasaje
secreto."""

entrada = input("Ingrese las palabras spoiler (separadas por coma): ")

spoilers = entrada.split(",") #convierte el texto en una lista de palabras

for i in range(len(spoilers)): #esto recorre la lista de spoilers y elimina espacios en blanco al inicio y al final de cada palabra
    spoilers[i] = spoilers[i].strip()

palabras = review.split() #divide el texto en palabras individuales
resultado = [] #aca se guardarán las palabras del texto final, reemplazando las palabras spoiler por asteriscos

for palabra in palabras:
    palabra_limpia = palabra.strip(".,;:¡!¿?()").lower() #elimina signos de puntuación y convierte a minúscula para comparar con la lista de spoilers
    reemplazada = palabra

    for spoiler in spoilers: #compara cada palabra limpia con cada palabra spoiler, si hay coincidencia reemplaza la palabra original por asteriscos
        if palabra_limpia == spoiler.lower():
            reemplazada = "*" * len(palabra_limpia)
    
    resultado.append(reemplazada)

texto_final = " ".join(resultado)
print(texto_final)
```
