Importación de funciones:

Estas funciones ya fueron definidas en el archivo `ejercicio1.py`, por lo que no hace falta reescribir la lógica en el notebook.

```python
import sys
import os
sys.path.append(os.path.abspath(".."))
from src.ejercicio1 import (
    obtener_lineas,
    contar_lineas,
    contar_palabras,
    calcular_promedio,
    lineas_mayores_al_promedio
)
```

Texto a analizar:
En esta celda definimos el texto que vamos a procesar.  
Luego, con las funciones importadas, podremos contar sus líneas, palabras y calcular el promedio de palabras por línea.

```python
texto = """Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.
Flat is better than nested.
Sparse is better than dense.
Readability counts.
Special cases aren't special enough to break the rules.
Although practicality beats purity.
Errors should never pass silently.
Unless explicitly silenced.
In the face of ambiguity, refuse the temptation to guess.
There should be one-- and preferably only one --obvious way to do it.
Although that way may not be obvious at first unless you're Dutch.
Now is better than never.
Although never is often better than *right* now.
If the implementation is hard to explain, it's a bad idea.
If the implementation is easy to explain, it may be a good idea.
Namespaces are one honking great idea -- let's do more of those!"""
```

Procesamiento del texto
En esta celda aplicamos las funciones del archivo `src/ejercicio1.py`.
Primero obtenemos la lista de líneas del texto.
Después contamos cuántas líneas hay.
Luego contamos la cantidad total de palabras.
Finalmente calculamos el promedio de palabras por línea.

```python
lineas = obtener_lineas(texto)
total_lineas = contar_lineas(lineas)
total_palabras = contar_palabras(lineas)
promedio = calcular_promedio(total_palabras, total_lineas)
```

Resultados obtenidos
En esta celda mostramos:
la cantidad total de líneas,
la cantidad total de palabras,
el promedio de palabras por línea,
y las líneas que tienen más palabras que ese promedio.

```python
print("Total de líneas:", total_lineas)
print("Total de palabras:", total_palabras)
print("Promedio de palabras por línea:", round(promedio, 2))

print("\nLíneas por encima del promedio:\n")

for linea in lineas_mayores_al_promedio(lineas, promedio):
    print("-", linea)
```
