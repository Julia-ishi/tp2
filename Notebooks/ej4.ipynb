```python
email = input("Ingrese un email: ")

valido = True

if email.count("@") != 1: #verifica que haya exactamente un símbolo @ en el email
    valido = False

if email.startswith("@") or email.endswith("@"): #verifica que el email no comience ni termine con el símbolo @
    valido = False

if email.startswith(".") or email.endswith("."):
    valido = False

if valido:
    partes = email.split("@") #divide el email en dos partes usando el símbolo @ como separador
    antes_arroba = partes[0] #la parte antes del símbolo @
    despues_arroba = partes[1]  #la parte después del símbolo @

    if len(antes_arroba) < 1: #verifica que la parte antes del símbolo @ tenga al menos un carácter
        valido = False

    if "." not in despues_arroba: #verifica que la parte después del símbolo @ contenga al menos un punto
        valido = False

    else:
        ultimo_punto = despues_arroba.rfind(".") #encuentra la posición del último punto en la parte después del símbolo @
        dominio = despues_arroba[ultimo_punto + 1:] #extrae la parte del dominio después del último punto

        if len(dominio) < 2:    #verifica que la parte del dominio tenga al menos dos caracteres (por ejemplo, "com", "org", etc.)
            valido = False

if valido:
    print("El email es válido.")
else:
    print("El email no es válido.")
```
