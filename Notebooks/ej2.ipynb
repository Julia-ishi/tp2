Importación de funciones
En esta celda importamos desde la carpeta `src` las funciones necesarias para resolver el ejercicio.
Estas funciones permiten:
convertir duraciones a segundos,
calcular la duración total de la playlist,
encontrar la canción más larga y más corta,
y convertir segundos a formato minutos y segundos.

```python
import sys
import os
sys.path.append(os.path.abspath(".."))
from src.ejercicio2 import (
    calcular_duracion_total,
    encontrar_cancion_mas_larga,
    encontrar_cancion_mas_corta,
    convertir_total_a_minutos_segundos
)
```

Definición de la playlist
En esta celda definimos una lista de canciones, donde cada canción tiene:
un título (`title`)
una duración en formato "mm:ss"

```python
playlist = [
    {"title": "Bohemian Rhapsody", "duration": "5:55"},
    {"title": "Hotel California", "duration": "6:30"},
    {"title": "Stairway to Heaven", "duration": "8:02"},
    {"title": "Imagine", "duration": "3:07"},
    {"title": "Smells Like Teen Spirit", "duration": "5:01"},
    {"title": "Billie Jean", "duration": "4:54"},
    {"title": "Hey Jude", "duration": "7:11"},
    {"title": "Like a Rolling Stone", "duration": "6:13"},
]
```

Procesamiento de la playlist
En esta celda:
calculamos la duración total de todas las canciones
convertimos ese total a minutos y segundos
encontramos la canción más larga y la canción más corta.

```python
total_segundos = calcular_duracion_total(playlist)
minutos_totales, segundos_restantes = convertir_total_a_minutos_segundos(total_segundos)
mas_larga = encontrar_cancion_mas_larga(playlist)
mas_corta = encontrar_cancion_mas_corta(playlist)
```

Resultados obtenidos
En esta celda mostramos:
la duración total de la playlist
la canción más larga y la canción más corta.

```python
print(f"Duración total: {minutos_totales}m {segundos_restantes}s")
print(f'Canción más larga: "{mas_larga[0]}" ({mas_larga[1] // 60}:{mas_larga[1] % 60:02d})')
print(f'Canción más corta: "{mas_corta[0]}" ({mas_corta[1] // 60}:{mas_corta[1] % 60:02d})')
```
